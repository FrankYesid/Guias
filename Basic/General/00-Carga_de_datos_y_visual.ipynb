### Carga de datos: Cargamos tanto el archivo mnist_train.csv como mnist_test.csv.


```python
# Importar pandas que usamos para la carga y manipulación de los datos en formato tabla (Estructurados)
import pandas as pd # Para trabajar con datos tabulares y archivos CSV xlsx, entre otros.

# Cargar los archivos CSV (cambia las rutas si es necesario)
# Asegúrate de que el archivo mnist.csv esté en la misma carpeta o proporciona la ruta completa que la del cuaderno o algoritmo
train_df = pd.read_csv('mnist_train.csv')  # Carga los datos de entrenamiento desde el archivo CSV.
test_df = pd.read_csv('mnist_test.csv')    # Carga los datos de prueba desde el archivo CSV.
```

```python
# Mostrar las primeras filas del DataFrame para revisar los datos
print("Primeras 5 filas del conjunto de entrenamiento:")
print(train_df.head())  # Muestra las primeras 5 filas del archivo de entrenamiento.
print("\nPrimeras 5 filas del conjunto de prueba:")
print(test_df.head())   # Muestra las primeras 5 filas del archivo de prueba.
```

```python

# Verificar la cantidad total de datos
print("Forma de los datos de entrenamiento (filas, columnas):")
print(train_df.shape)  # Imprime la cantidad de filas y columnas del archivo de entrenamiento.
print("\nForma de los datos de prueba (filas, columnas):")
print(test_df.shape)   # Imprime la cantidad de filas y columnas del archivo de prueba.
```

```python

# Verificar la cantidad de valores únicos por cada campo
print("Valores únicos por columna en los datos de entrenamiento:")
print(train_df.nunique())  # Imprime el número de valores únicos por cada columna en el archivo de entrenamiento.
print("\nValores únicos por columna en los datos de prueba:")
print(test_df.nunique())   # Imprime el número de valores únicos por cada columna en el archivo de prueba.
```

```python

# Verificar si existen datos vacíos (NaN)
print("Datos faltantes en los datos de entrenamiento:")
print(train_df.isnull().sum())  # Verifica si hay valores faltantes (NaN) en el archivo de entrenamiento.
print("\nDatos faltantes en los datos de prueba:")
print(test_df.isnull().sum())   # Verifica si hay valores faltantes (NaN) en el archivo de prueba.
```

```python
# Resumen estadístico básico de los datos
print("\nResumen estadístico de los datos:")
print(train_df.describe())
```

```python
# Función para verificar los campos en un DataFrame
def verificar_campos(df, nombre):
    print(f"\n--- Verificación para {nombre} ---")
    
    # Mostrar primeras 5 filas
    print(f"\nPrimeras 5 filas de {nombre}:")
    print(df.head())
    
    # Mostrar el número total de filas y columnas
    print(f"\nCantidad total de datos (filas, columnas) en {nombre}:")
    print(df.shape)
    
    # Verificar la cantidad de valores únicos por cada campo
    print(f"\nValores únicos por campo en {nombre}:")
    print(df.nunique())
    
    # Verificar si existen datos vacíos (NaN)
    print(f"\nDatos faltantes en {nombre}:")
    print(df.isnull().sum())
    
    # Resumen estadístico básico
    print(f"\nResumen estadístico para {nombre}:")
    print(df.describe())

# Llamar a la función para verificar los archivos de entrenamiento y prueba
verificar_campos(train_df, 'mnist_train.csv')
verificar_campos(test_df, 'mnist_test.csv')
```

### Visualización  de datos

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt # Para generar gráficos en Python.
import seaborn as sns           # Para generar gráficos con estilo basado en matplotlib.
```

```python
# Graficar un histograma de los datos de entrenamiento
plt.figure(figsize=(10,6))   # Define el tamaño de la figura.
train_df.iloc[:, 1:3].hist(bins=20, figsize=(16,12), color='blue', alpha=0.7)  
plt.suptitle("Histograma de los píxeles en mnist_train.csv", fontsize=16)
plt.show()
```

train_df.iloc[:, 1:2]: Seleccionamos todas las columnas menos la primera (label), ya que las otras columnas son los valores de los píxeles.\
.hist(): Se crea un histograma que muestra la distribución de los valores de los píxeles.\
bins=20: Divide los datos en 20 grupos o intervalos.\
color='blue' y alpha=0.7: Configura el color de las barras del histograma y la transparencia (0.7 es parcialmente transparente).

```python
# Graficar un gráfico de barras de las etiquetas de los datos de entrenamiento

plt.figure(figsize=(8,6))   # Define el tamaño de la figura.
sns.countplot(train_df.iloc[:10]['label'], palette='Set2')  # Grafica la cantidad de ejemplos de cada dígito.
plt.title("Distribución de las etiquetas en mnist_train.csv")  # Título del gráfico.
plt.xlabel('Etiqueta (Dígito)')  # Etiqueta en el eje x.
plt.ylabel('Cantidad')           # Etiqueta en el eje y.
plt.legend(train_df.iloc[:10]['label'])
plt.show()
```

sns.countplot(): Crea un gráfico de barras que muestra cuántos ejemplos hay de cada etiqueta (dígito).\
El eje X representa las etiquetas (0-9) y el eje Y el número de imágenes correspondientes a cada etiqueta.

```python
# Graficar un gráfico de líneas de los valores promedio de los píxeles
plt.figure(figsize=(10,6))   # Define el tamaño de la figura.
avg_pixels = train_df.iloc[:, 1:].mean(axis=0)  # Calcula el promedio de los valores de cada píxel a lo largo de todas las imágenes.
plt.plot(avg_pixels, color='red')  # Grafica el promedio de los píxeles.
plt.title("Promedio de los valores de los píxeles en mnist_train.csv")  # Título del gráfico.
plt.xlabel('Índice del píxel')   # Etiqueta del eje x.
plt.ylabel('Valor promedio del píxel')   # Etiqueta del eje y.
plt.show()
```

Explicación:

.mean(axis=0): Calcula el valor promedio de cada columna (píxel) en todas las imágenes.\
plt.plot(): Crea un gráfico de líneas para visualizar el promedio de los valores de los píxeles.\
El gráfico nos da una idea de qué tan "brillantes" son los píxeles en promedio.

```python
# Visualización de ejemplos de imágenes de dígitos
plt.figure(figsize=(10,10))   # Define el tamaño de la figura.

sample = train_df.sample(9)   # Selecciona 9 imágenes aleatorias del conjunto de entrenamiento.

# Iterar sobre las 9 muestras seleccionadas
for i in range(9):
    img = sample.iloc[i, 1:].values.reshape(28, 28)  # Reorganiza las 784 columnas de píxeles en una imagen de 28x28.
    label = sample.iloc[i, 0]  # La primera columna es la etiqueta (el dígito).

    plt.subplot(3, 3, i+1)   # Crea una cuadrícula de 3x3 para mostrar las 9 imágenes.
    plt.imshow(img, cmap='gray')  # Muestra la imagen en escala de grises.
    plt.title(f'Label: {label}')  # Muestra la etiqueta de la imagen.
    plt.axis('off')   # Oculta los ejes.

plt.suptitle("Ejemplos de dígitos en mnist_train.csv", fontsize=16)  # Título general del gráfico.
plt.tight_layout()   # Ajusta el diseño de la cuadrícula para que no se superpongan las imágenes.
plt.show()
```

sample(): Selecciona aleatoriamente 9 ejemplos del conjunto de datos.\
.reshape(28, 28): Reorganiza los 784 valores de los píxeles en una imagen de 28x28.\
plt.subplot(3, 3, i+1): Crea una cuadrícula de 3x3 para mostrar las imágenes.\
plt.imshow(): Visualiza la imagen en escala de grises.\
cmap='gray': Utiliza una escala de color en grises para visualizar los píxeles.

```python

# Visualización de las clases (números) en el conjunto de datos de entrenamiento
plt.figure(figsize=(10,6))
sns.countplot(train_df['label'])
plt.title('Distribución de las etiquetas (dígitos) en el conjunto de datos MNIST en entrenamiento')
plt.show()

# Visualización de algunos ejemplos de dígitos en el conjunto de datos
def plot_digit_samples(df, num_samples=9):
    """
    Función para visualizar muestras de los dígitos
    """
    # Seleccionar una muestra aleatoria del conjunto de datos
    sample = df.sample(num_samples)
    
    # Configurar la figura para mostrar 9 ejemplos
    plt.figure(figsize=(10,10))
    
    # Iterar sobre la muestra para graficar cada imagen
    for i in range(num_samples):
        img = sample.iloc[i, 1:].values.reshape(28, 28)  # Las imágenes son de 28x28
        label = sample.iloc[i, 0]  # La primera columna es la etiqueta 'label'
        
        plt.subplot(3, 3, i+1)
        plt.imshow(img, cmap='gray')
        plt.title(f'Label: {label}')
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Llamar a la función para mostrar muestras de los dígitos
plot_digit_samples(train_df)
```

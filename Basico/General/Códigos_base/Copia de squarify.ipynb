```
import squarify
import matplotlib.pyplot as plt

labels=['nepal', 'america', 'india']
sizes=[2, 3, 4]
colors=['red', 'blue', 'red']

squarify.plot(sizes=sizes, 
             label=labels, 
             color =colors,
             alpha=.7,
             bar_kwargs=dict(linewidth=1, edgecolor="#222222"))
plt.show()
```

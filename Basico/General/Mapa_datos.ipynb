```python
#@title Mapa
###@title Organizar departamentos con latitud y longitud { display-mode: "form" }
deptos= df.REGION[df.REGION != 'IGNORADA'][df.REGION != ''].value_counts().index.tolist()#tmp_[tmp_.contactado_==0].LOCALIDAD.value_counts().index.tolist()
deptos[0]  = '13-METROPOLITANA DE SANTIAGO CHILE'
deptos[1]  = '05-VALPARAISO CHILE'
deptos[4]  = '08-BIO BIO CHILE'
deptos[5]  = '09-LA ARAUCANIA CHILE'
deptos[6]  = '02-ANTOFAGASTA CHILE'
deptos[7]  = '07-MAULE CHILE'

deptos1 = pd.DataFrame()
deptos1['Departamentos'] = df.REGION[df.REGION != 'IGNORADA'][df.REGION != ''].value_counts().index.tolist() #tmp_[tmp_[meses].sum(axis=1)>0].LOCALIDAD.value_counts().index.tolist()
deptos1['Valores']       =  df.REGION[df.REGION != 'IGNORADA'][df.REGION != ''].value_counts().values.tolist()#tmp_[tmp_[meses].sum(axis=1)>0].LOCALIDAD.value_counts().values.tolist()
deptos1['Departamentos'].iloc[0]  = '13-METROPOLITANA DE SANTIAGO CHILE'
deptos1['Departamentos'].iloc[1]  = '05-VALPARAISO CHILE'
deptos1['Departamentos'].iloc[4]  = '08-BIO BIO CHILE'
deptos1['Departamentos'].iloc[5]  = '09-LA ARAUCANIA CHILE'
deptos1['Departamentos'].iloc[6]  = '02-ANTOFAGASTA CHILE'
deptos1['Departamentos'].iloc[7]  = '07-MAULE CHILE'

from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='cop')

lat = []
lon = []
for i in range(len(deptos)):
  location = geolocator.geocode(deptos1.loc[i,'Departamentos'])
  print(deptos[i])
  print(location)
  lat.append(location.latitude)
  lon.append(location.longitude)

deptos1['Latitud'] = lat
deptos1['Longitud'] = lon

deptos1['Departamentos'].iloc[0]  = '13-METROPOLITANA DE SANTIAGO'
deptos1['Departamentos'].iloc[1]  = '05-VALPARAISO'
deptos1['Departamentos'].iloc[4]  = '08-BIO BIO'
deptos1['Departamentos'].iloc[5]  = '09-LA ARAUCANIA'
deptos1['Departamentos'].iloc[6]  = '02-ANTOFAGASTA'
deptos1['Departamentos'].iloc[7]  = '07-MAULE'
```

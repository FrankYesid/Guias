# Basic Visualization with Matplotlib

Brief explanation of Matplotlib and its usage.
Create basic plots such as line, scatter, bar, and pie charts

```python
# Line plot example
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sine')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Advanced Visualization with Matplotlib

Show more advanced features like subplots, plot styles, and axis customization.
Example with multiple subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y, 'r')
axs[0, 0].set_title('Subplot 1')

axs[0, 1].bar([1, 2, 3], [3, 5, 7], color='b')
axs[0, 1].set_title('Subplot 2')

axs[1, 0].scatter(np.random.rand(10), np.random.rand(10))
axs[1, 0].set_title('Subplot 3')

plt.tight_layout()
plt.show()
```

# Visualization with Seaborn

Introduction to Seaborn as a tool for statistical visualization.
Create charts like distributions, box plots, heatmaps, and regression plots

```python
# Load example dataset
data = sns.load_dataset('tips')

# Distribution plot
plt.figure(figsize=(8, 5))
sns.histplot(data['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.show()

# Box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='day', y='total_bill', data=data)
plt.title('Boxplot of Bills by Day')
plt.show()

# Heatmap
corr = data.corr()
plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Graph Visualization with NetworkX

Introduction to graph visualization using NetworkX.
Create and visualize simple graphs and show how to customize nodes and edges.

```python
# Create a simple graph
G = nx.Graph()

# Add nodes and edges
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Draw the graph
plt.figure(figsize=(8, 5))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=700, font_size=15, edge_color='gray')
plt.title('Simple Graph Visualization')
plt.show()
```

#Neural Network Visualization with NetworkX

In this section, we will visualize a simple neural network structure using NetworkX. Each layer of the neural network will be represented as a set of nodes, and the connections between the layers as edges.

Neural Network Structure Example:
- Input Layer: 3 neurons

- Hidden Layer: 4 neurons

- Output Layer: 2 neurons

We will visually represent the flow of information from the input layer through the hidden layer and to the output layer.

```python
# Define the layers of the neural network
input_layer = [f"Input {i+1}" for i in range(3)]
hidden_layer = [f"Hidden {i+1}" for i in range(4)]
output_layer = [f"Output {i+1}" for i in range(2)]

# Create a graph
G = nx.DiGraph()  # Directed graph to show flow direction

# Add nodes for each layer
G.add_nodes_from(input_layer, layer='Input')
G.add_nodes_from(hidden_layer, layer='Hidden')
G.add_nodes_from(output_layer, layer='Output')

# Add edges (connections) between layers
for input_neuron in input_layer:
    for hidden_neuron in hidden_layer:
        G.add_edge(input_neuron, hidden_neuron)

for hidden_neuron in hidden_layer:
    for output_neuron in output_layer:
        G.add_edge(hidden_neuron, output_neuron)

# Set positions for each layer
pos = {}

# Spread input layer at the bottom
for i, neuron in enumerate(input_layer):
    pos[neuron] = (i, 0)

# Spread hidden layer in the middle
for i, neuron in enumerate(hidden_layer):
    pos[neuron] = (i, 1)

# Spread output layer at the top
for i, neuron in enumerate(output_layer):
    pos[neuron] = (i, 2)

# Draw the network
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightgreen', font_size=12, edge_color='gray')
plt.title('Neural Network Configuration Visualization')
plt.show()
```

Explanation:
Nodes: Each node represents a neuron in a specific layer (input, hidden, output).
Edges: Directed edges represent the flow of data between neurons in the network.
Layout: The positions of the nodes are organized by layer, using a manual layout to visualize the structure clearly.

# Customizing Neural Network Visualization:

If you'd like to add more details, such as weights between neurons or activations, you can label the edges or change the edge thickness to represent weights:

```python
# Assign random weights to edges and visualize them
weights = np.random.rand(len(G.edges))

# Draw the network with edge weights
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=12, edge_color='gray')
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f'{w:.2f}' for (u, v), w in zip(G.edges, weights)}, font_color='red')
plt.title('Neural Network with Edge Weights')
plt.show()
```
